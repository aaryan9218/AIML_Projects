```python
import pandas as pd
import numpy 

df=pd.read_csv(r'C:\Users\adash\Downloads\popularity.csv')
df1=df['shares']

elements = numpy.array(df1)

print('length before:',len(elements))      
mean = numpy.mean(elements, axis=0)
sd = numpy.std(elements, axis=0)


print('mean before:',mean)
final_list = [x for x in df1 if (x > mean - 1.5 * sd)]
final_list = [x for x in final_list if (x < mean + 1.5 * sd)]

print('length after',len(final_list))
      
print('mean after:',sum(final_list)/len(final_list))

```
