```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\adash\Downloads\odi-batting.csv')

df['Century'] = df['Runs'].apply(lambda x : 1 if x>=100 else 0)

df1=df[['Player','Century']].groupby(['Player']).sum()

print (df1[(df1.Century==df1.Century.max())])
 
```
